### Introdução

Hoje, há um problema de considerável tamanho no ambiente médico-hospitalar em que médicos pedem equipamentos ou medicamentos em demasia aos operadores de planos de saúde. Os equipamentos podem acabar não sendo utilizados, e os medicamentos podem acabar sendo repassados ao paciente sem a necessidade (o que é potencialmente mais arriscado, pois pode induzir o paciente a uma overdose de medicamentos). 

Com o objetivo de reduzir este desperdício, o grupo criou neste Hackaturing uma maneira de as operadoras de planos de saúde terem informações a respeito de, com base no código do item requisitado e no tipo de tratamento que está sendo feito, elas saberem se é mais provável que o pedido seja pago em sua totalidade ou não.

Isso permite às empresas saber quais famílias de produtos possuem as maiores diferenças entre valor total requisitado e valor efetivamente pago. Um delta alto entre esses valores indica que (i) os médicos estão requisitando um determinado produto em demasia ou (ii) os planos de saúde possuem políticas inadequadas de liberação desses itens que podem ser bastante necessários para o médico e o paciente. 

De qualquer maneira, essa inteligência de dados ajudará as empresas de planos de saúde reduzirem ineficiências e tomarem melhores decisões.

### 3. Modelling

Para a parte de modelagem, vamos tentar diferentes classifiers e observar quais oferecem a maior acurácia possível.

```python
df_crm = df['base_hackaturing.crm_solicitante']
df = df.drop(['base_hackaturing.crm_solicitante'], axis=1)
df.head()
```

```python
df_zero = df[df['base_hackaturing.pago_ou_nao'] == 0]
df_one = df[df['base_hackaturing.pago_ou_nao'] == 1]

df_one = df_one.sample(n=15000)
```

```python
export_csv = df.to_csv(r'C:\Users\Bruno Koba\Desktop\final_dataframe.csv', index = None, header=True) 
```

```python
# Dividing dataset into X and y portions

import numpy as np
from sklearn.model_selection import train_test_split

df = pd.concat([df_zero, df_one])

X = df.iloc[:, :-1]  
y = df.iloc[:, -1]

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.20, random_state=42)
```

```python
# Training and evaluating 5 differente classifiers, and saving the best performer as best_classifier

from sklearn.metrics import accuracy_score, log_loss
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC, LinearSVC, NuSVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier, GradientBoostingClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
import matplotlib.pyplot as plt

classifiers = [
    KNeighborsClassifier(3),
    RandomForestClassifier(),
    AdaBoostClassifier(),
    GradientBoostingClassifier(),
    LinearDiscriminantAnalysis()]
    
best_classifier = 0
record_accuracy = 0

# Logging for Visual Comparison
log_cols=["Classifier", "Accuracy", "Log Loss"]
log = pd.DataFrame(columns=log_cols)

for clf in classifiers:
    clf.fit(X_train, y_train)
    name = clf.__class__.__name__
    
    print("="*30)
    print(name)
    
    print('****Results****')
    train_predictions = clf.predict(X_test)
    acc = accuracy_score(y_test, train_predictions)
    print("Accuracy: {:.4%}".format(acc))
    if acc > record_accuracy:
        record_accuracy = acc
        best_classifier = clf
    
    train_predictions = clf.predict_proba(X_test)
    ll = log_loss(y_test, train_predictions)
    print("Log Loss: {}".format(ll))
    
    log_entry = pd.DataFrame([[name, acc*100, ll]], columns=log_cols)
    log = log.append(log_entry)
    
sns.set_color_codes("muted")
sns.barplot(x='Accuracy', y='Classifier', data=log, color="b")

plt.xlabel('Accuracy %')
plt.title('Classifier Accuracy')
plt.show()

sns.set_color_codes("muted")
sns.barplot(x='Log Loss', y='Classifier', data=log, color="g")

plt.xlabel('Log Loss')
plt.title('Classifier Log Loss')
plt.show()
```

```python
from sklearn.metrics import classification_report

best_classifier.fit(X_train, y_train)
y_pred = best_classifier.predict(X_test)

target_names = ['nao_pago', 'pago']
print(classification_report(y_test, y_pred, target_names=target_names))
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred)
```

```python
import numpy as np
from sklearn.metrics import roc_auc_score

roc_auc_score(y_test, y_pred)
```
